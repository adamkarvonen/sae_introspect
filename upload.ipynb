```python


def upload_acts_to_hf():
    from huggingface_hub import HfApi

    api = HfApi()

    filename = "max_acts/acts_google_gemma-2-9b-it_layer_9_trainer_131_layer_percent_25_context_length_32.pt"

    api.upload_file(
        path_or_fileobj=filename,
        path_in_repo=filename.split("/")[-1],
        repo_type="dataset",
        repo_id="adamkarvonen/sae_max_acts",
    )


upload_acts_to_hf()
```
